```python
data = pd.read_csv("ai-training-computation.csv").query("year > 1981")

base = alt.Chart(data).mark_point(clip=True).encode(
    alt.Y('Training_computation_petaflop', scale=alt.Scale(type='log')),
    alt.X('year', scale=alt.Scale(domain=(1985, 2022))),
    color='Domain'
)
polynomial_fit = [
    base.transform_regression(
        "year", "Training_computation_petaflop", method="exp", extent=[1985, 2022]#, as_=["year", "used-to-be-order"]
    )
    .mark_line()
]
alt.layer(base, *polynomial_fit)
```

```python

data = pd.read_csv("attendance-major-ai-conferences.csv").query("Entity == 'Total'")

base = alt.Chart(data).mark_bar().encode(
    alt.Y('conference_attendance', scale=alt.Scale()),
    alt.X('Year', scale=alt.Scale(zero=False)),
    color='Entity'
)



line = alt.Chart(data).mark_line(color='red').transform_window(
    rolling_mean='mean(conference_attendance)',
).encode(
    alt.X('Year', scale=alt.Scale(zero=False)),
    y='rolling_mean:Q'
)
(base + line).properties(width=600)

polynomial_fit = base.transform_regression(
    "Year", "conference_attendance", method="exp", #extent=[1985, 2022]#, as_=["year", "used-to-be-order"]
).mark_line()

alt.layer(base, polynomial_fit)

```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
# 0 -> neither pursue nor possessed
# 1 -> allegedly pursed
# 2 -> pursed
# 3 -> a possessed
# 4 -> possessed
# 5 -> used

# Note: not historical.
data = pd.read_csv("../biotech/historical-biological-weapons.csv")
display(data.head)
data.columns
# TODO: probably requires a different data visualization library for world map visualization

import altair as alt
from vega_datasets import data

# Data generators for the background
sphere = alt.sphere()
graticule = alt.graticule()

# Source of land data
source = alt.topo_feature(data.world_110m.url, 'countries')

# Layering and configuring the components
alt.layer(
    alt.Chart(sphere).mark_geoshape(fill='lightblue'),
    alt.Chart(graticule).mark_geoshape(stroke='white', strokeWidth=0.5),
    alt.Chart(source).mark_geoshape(fill='ForestGreen', stroke='black')
).project(
    'naturalEarth1'
).properties(width=600, height=400).configure_view(stroke=None)
```
